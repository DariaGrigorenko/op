```python
import time
import random

class HashTable:
    def __init__(self, size):
        self.size = size
        self.slots = [None] * self.size
        self.count = 0 # колво

    def hash_function(self, key): #index
        return key % self.size

    def rehash(self, old_hash, step_size=1):
        return (old_hash + step_size) % self.size
    
    def is_full(self):
        return self.count == self.size
    
    def add(self, key, gen):
        if self.is_full():
            print("Хеш-таблица переполнена")
            return
        
        hash_value = self.hash_function(key) #выч хэш

        if self.slots[hash_value] is None:
            self.slots[hash_value] = (key, gen)
            self.count += 1
        else:
            next_slot = self.rehash(hash_value)
            while self.slots[next_slot] is not None and self.slots[next_slot][0] != key:
                next_slot = self.rehash(next_slot)

            if self.slots[next_slot] is None:
                self.slots[next_slot] = (key, gen)
                self.count += 1
            else:
                self.slots[next_slot] = (key, gen)
                print(f"Значение по ключу {key} обновлено.")

    def get(self, key):
        start_slot = self.hash_function(key)

        x = 'Значение не найдено'
        stop = False
        found = False #найден
        position = start_slot

        while self.slots[position] is not None and not found and not stop:
            if self.slots[position][0] == key:
                found = True
                x = self.slots[position][1]
            else:
                position = self.rehash(position)
                if position == start_slot:
                    stop = True

        return x

def remove(self, key):
        start_slot = self.hash_function(key)
        position = start_slot

        while self.slots[position] is not None:
            if self.slots[position][0] == key:
                self.slots[position] = None
                self.count -= 1
                print(f"Элемент с ключом {key} удалён.")
                return
            position = self.rehash(position)
            if position == start_slot:
                break
        print("Элемент не найден для удаления.")

Size = int(input("Введите размер хэш таблицы: "))
h_table = HashTable(Size) 
while True:
    command = input("Введите команду (add/get/remove/show/exit): ")
    if command == 'add': 
        keyP = int(input("Введите ключ: "))
        dataP = input("Введите значение: ")
        h_table.add(keyP, dataP)
    elif command == 'get': 
        keyG = int(input("Введите ключ для поиска: "))
        start_time = time.time()
        print(h_table.get(keyG))
        print("--- {0} ms ---".format(round((time.time() - start_time) * 1000)))
    elif command == 'remove':
        key = int(input("Введите ключ для удаления: "))
        h_table.remove(key)
    elif command == 'show': # Отображение хэш таблицы
        print(h_table.slots)
    elif command == 'exit': # Завершение программы
        break
    else:
        print("Неверная команда. Попробуйте снова.")
```

```python
import time 
import random


class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for i in range(size)]

    def hash_function(self, key):
        return key % self.size ## index
    
    def add(self, key, value):
        hash_value = self.hash_function(key) #выч хэш
        self.table[hash_value].append((key, value)) 

    def remove(self, key):
        hash_value = self.hash_function(key)
        for i, (k, v) in enumerate(self.table[hash_value]):
            if k == key:
                del self.table[hash_value][i]  
                print(f"Элемент с ключом {key} удалён.")
                return
        print(f"Элемент с ключом {key} не найден для удаления.")

    def get(self, key):
        start_slot = self.hash_function(key)

        for k1, k2 in self.table[start_slot]:
            if k1 == key:
                return k2
        return 'значение не найдено'

Size = int(input("Введите размер хэш таблицы: "))
h_table = HashTable(Size) 
while True:
    command = input("Введите команду (add/get/remove/show/exit): ")
    if command == 'add': 
        keyP = int(input("Введите ключ: "))
        dataP = input("Введите значение: ")
        h_table.add(keyP, dataP)
    elif command == 'get': 
        keyG = int(input("Введите ключ для поиска: "))
        start_time = time.time()
        print(h_table.get(keyG))
        print("--- {0} ms ---".format(round((time.time() - start_time) * 1000)))
    elif command == 'remove':
        keyR = int(input("Введите ключ для удаления: "))
        h_table.remove(keyR)
    elif command == 'show': # Отображение хэш таблицы
        print(h_table.table)
    elif command == 'exit': # Завершение программы
        break
    else:
        print("Неверная команда. Попробуйте снова.")
```

### Задание №3
Расставить на стандартной 64-клеточной шахматной доске 8 ферзей так, чтобы ни один из них не находился под боем другого». Подразумевается, что ферзь бьёт все клетки, расположенные по вертикалям, горизонталям и обеим диагоналям
Написать программу,  которая находит хотя бы один способ решения задач.

```python
def is_valid(row, col, board): 
    # проверка, бьются ли ферзи в той же строке или диагоналях 
    for i in range(row): 
        if board[i] == col or board[i] - i == col - row or board[i] + i == col + row: 
            return False 
    return True 
def solve(size, row, board, results): 
    # если все столбцы заполнены, добавляем расстановку ферзей в результаты 
    if row == size: 
        results.append(board.copy()) 
        return 
    # пытаемся поставить ферзя в каждую строку текущего столбца 
    for col in range(size): 
        if is_valid(row, col, board): 
            board[row] = col 
            solve(size, row+1, board, results) 
            board[row] = -1 
def find_all_solutions(size): 
    results = [] 
    board = [-1] * size #доска
    solve(size, 0, board, results) 
    return results 

def print_solution(solution):
    size = len(solution)
    for row in range(size):
        line = ""
        for col in range(size):
            if solution[row] == col:
                line += "Q "
            else:
                line += ". "
        print(line)
    print()

all_solutions = find_all_solutions(8) 
print(f"Найдено {len(all_solutions)} решений для доски 8x8")
for i, solution in enumerate(all_solutions[:2], 1):
    print(f"Решение #{i}:")
    print_solution(solution)
```
