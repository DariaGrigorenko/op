```python
import random
import time

# –ì–µ–Ω–µ—Ä–∞—Ü–∏—è
gen = [random.randint(1, 1000) for i in range(1000)]
newgen = sorted(gen)

def update_sorted():
    global newgen
    newgen = sorted(gen)

def plus(var):
    gen.append(var)
    update_sorted()
    print(f"–≠–ª–µ–º–µ–Ω—Ç {var} –¥–æ–±–∞–≤–ª–µ–Ω.")

def deli(var):
    if var in gen:
        gen.remove(var)
        update_sorted()
        print(f"–≠–ª–µ–º–µ–Ω—Ç {var} —É–¥–∞–ª—ë–Ω.")
    else:
        print(f"–≠–ª–µ–º–µ–Ω—Ç {var} –Ω–µ –Ω–∞–π–¥–µ–Ω.")

def bin_poick(var):
    left = 0 
    right = len(newgen) - 1
    while left <= right:
        middle = (left + right) // 2
        if newgen[middle] == var:
            return middle
        elif newgen[middle] < var:
            left = middle + 1
        else:
            right = middle - 1
    
    return -1
#Num1 = int(input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ: '))
#bin_poick(Num1)
#ranNum = random.choice(gen)  
#res1 = bin_poick(ranNum)

class bin_tree:
    def __init__(self, key):
        self.left = None
        self.right = None   
        self.val = key

def insert(root, key):
    if root is None:
        return bin_tree(key)
    else:
        if root.val < key:
            root.right = insert(root.right, key)
        else:
            root.left = insert(root.left, key)
    return root

def tree_search(root, key):
    if root is None or root.val == key:
        return root
    if root.val < key:
        return tree_search(root.right, key)
    return tree_search(root.left, key)

# –°–æ–∑–¥–∞–Ω–∏–µ –±–∏–Ω–∞—Ä–Ω–æ–≥–æ –¥–µ—Ä–µ–≤–∞ –ø–æ–∏—Å–∫–∞ –∏–∑ –º–∞—Å—Å–∏–≤–∞ 
def create_bin():
    root = None
    for val in gen:
        root = insert(root, val)
    return root

#root = create_bin()
#Num2 = int(input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ: '))
#res2 = tree_search(root, Num2)
#if res2 is not None:
#    print(f"–ß–∏—Å–ª–æ {Num2} –Ω–∞–π–¥–µ–Ω–æ –≤ –¥–µ—Ä–µ–≤–µ!")
#else:
 #   print(f"–ß–∏—Å–ª–æ {Num2} –Ω–µ –Ω–∞–π–¥–µ–Ω–æ –≤ –¥–µ—Ä–µ–≤–µ.")

def fibonacci(var):
    n = len(newgen)
    fib2 = 0
    fib1 = 1
    fib = fib1 + fib2
    while fib < n:
        fib2 = fib1
        fib1 = fib
        fib = fib1 + fib2
    offset = -1
    while fib > 1:
        i = min(offset + fib2, n - 1)

        if newgen[i] < var:
            fib = fib1
            fib1 = fib2
            fib2 = fib - fib1
            offset = i
        elif newgen[i] > var:
            fib = fib2
            fib1 = fib1 - fib2
            fib2 = fib - fib1
        else:
            return i
    return -1
#Num4 = int(input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ: '))
#res4 = fibonacci(Num4)
#print(Num4, res4)

def inter(var): # 6
    low = 0
    high = len(newgen) - 1
    while low <= high and newgen[low] <= var <= newgen[high]:
        pos = low + ((var - newgen[low]) * (high - low)) // (newgen[high] - newgen[low])
        # 0 + ((6 - 1) * ((10-1) -0)) // (10 - 1)
        if newgen[pos] == var:
            return pos
        elif newgen[pos] < var:
            low = pos + 1
        else:
            high = pos - 1
    return -1

#Num5 = int(input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ: '))
#res5 = inter(Num5)
#print(Num5, res5)

while True:
    command = input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–º–∞–Ω–¥—É (add/remove/find/exit): ")

    if command == 'add': # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–∞
        var = int(input("–í–≤–µ–¥–∏—Ç–µ —ç–ª–µ–º–µ–Ω—Ç –¥–ª—è –¥–æ–±–∞–≤–ª–µ–Ω–∏—è: "))
        plus(var)
    elif command == 'remove': # –£–¥–∞–ª–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–∞
        var = int(input("–í–≤–µ–¥–∏—Ç–µ —ç–ª–µ–º–µ–Ω—Ç –¥–ª—è —É–¥–∞–ª–µ–Ω–∏—è: "))
        deli(var)

    elif command == 'find': # –ü–æ–∏—Å–∫ —ç–ª–µ–º–µ–Ω—Ç–∞
        command_find = input("–í–≤–µ–¥–∏—Ç–µ –Ω–æ–º–µ—Ä —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ–∏—Å–∫–∞ (–ë–∏–Ω–∞—Ä–Ω—ã–π –ø–æ–∏—Å–∫ (1) /–ë–∏–Ω–∞—Ä–Ω–æ–µ –¥–µ—Ä–µ–≤–æ (2) /–§–∏–±–æ–Ω–∞—á—á–∏–µ–≤ (3) /–ò–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏–æ–Ω–Ω—ã–π (4)): ")
        if command_find == '1':
            var = int(input("–í–≤–µ–¥–∏—Ç–µ —ç–ª–µ–º–µ–Ω—Ç –¥–ª—è –ø–æ–∏—Å–∫–∞: "))
            start_time = time.time()
            res = bin_poick(var)
            if res:
                print(f"–≠–ª–µ–º–µ–Ω—Ç {var} –Ω–∞–π–¥–µ–Ω –≤ –±–∏–Ω–∞—Ä–Ω–æ–º –¥–µ—Ä–µ–≤–µ –ø–æ–∏—Å–∫–∞.")
            else:
                print(f"–≠–ª–µ–º–µ–Ω—Ç {var} –Ω–µ –Ω–∞–π–¥–µ–Ω –≤ –±–∏–Ω–∞—Ä–Ω–æ–º –¥–µ—Ä–µ–≤–µ –ø–æ–∏—Å–∫–∞.")
            print("--- {0} ms ---".format(round((time.time() - start_time) * 1000)))
        elif command_find == '2':
            var = int(input("–í–≤–µ–¥–∏—Ç–µ —ç–ª–µ–º–µ–Ω—Ç –¥–ª—è –ø–æ–∏—Å–∫–∞: "))
            start_time = time.time()
            root = create_bin() # –°–æ–∑–¥–∞–Ω–∏–µ –±–∏–Ω–∞—Ä–Ω–æ–≥–æ –¥–µ—Ä–µ–≤–∞
            result = tree_search(root, var)
            if result:
                print(f"–≠–ª–µ–º–µ–Ω—Ç {var} –Ω–∞–π–¥–µ–Ω –≤ –±–∏–Ω–∞—Ä–Ω–æ–º –ø–æ–∏—Å–∫e.")
            else:
                print(f"–≠–ª–µ–º–µ–Ω—Ç {var} –Ω–µ –Ω–∞–π–¥–µ–Ω –≤ –±–∏–Ω–∞—Ä–Ω–æ–º –ø–æ–∏—Å–∫e.")
            print("--- {0} ms ---".format(round((time.time() - start_time) * 1000)))
        elif command_find == '3':
            var = int(input("–í–≤–µ–¥–∏—Ç–µ —ç–ª–µ–º–µ–Ω—Ç –¥–ª—è –ø–æ–∏—Å–∫–∞: "))
            start_time = time.time()
            r = fibonacci(var)
            if r != -1:
                print(f"–§–∏–±–æ–Ω–∞—á—á–∏–µ–≤ –ø–æ–∏—Å–∫: —ç–ª–µ–º–µ–Ω—Ç {var} –Ω–∞–π–¥–µ–Ω –Ω–∞ –ø–æ–∑–∏—Ü–∏–∏ {res}.")
            else:
                print(f"–§–∏–±–æ–Ω–∞—á—á–∏–µ–≤ –ø–æ–∏—Å–∫: —ç–ª–µ–º–µ–Ω—Ç {var} –Ω–µ –Ω–∞–π–¥–µ–Ω.")
            print("--- {0} ms ---".format(round((time.time() - start_time) * 1000)))
        elif command_find == '4':
            var = int(input("–í–≤–µ–¥–∏—Ç–µ —ç–ª–µ–º–µ–Ω—Ç –¥–ª—è –ø–æ–∏—Å–∫–∞: "))
            start_time = time.time()
            inter(var)
            print("--- {0} ms ---".format(round((time.time() - start_time) * 1000)))
            if var in gen:
                Index = sorted(gen).index(var)
                print(f"–≠–ª–µ–º–µ–Ω—Ç {var} –Ω–∞–π–¥–µ–Ω –≤ –ø–æ–∑–∏—Ü–∏–∏ {Index}.")
            else:
                print(f"–≠–ª–µ–º–µ–Ω—Ç {var} –Ω–µ –Ω–∞–π–¥–µ–Ω –≤ –º–∞—Å—Å–∏–≤–µ.")
            print("--- {0} ms ---".format(round((time.time() - start_time) * 1000)))
        else:
            print("–ù–µ–≤–µ—Ä–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –ø–æ–∏—Å–∫–∞. –ü–æ–ø—Ä–æ–±—É–π—Ç–µ —Å–Ω–æ–≤–∞.")

    elif command == 'exit': # –ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ –ø—Ä–æ–≥—Ä–∞–º–º—ã
        break
    else:
        print("–ù–µ–≤–µ—Ä–Ω–∞—è –∫–æ–º–∞–Ω–¥–∞. –ü–æ–ø—Ä–æ–±—É–π—Ç–µ —Å–Ω–æ–≤–∞.")
```

### –ó–∞–¥–∞–Ω–∏–µ ‚Ññ2
<i> 1 –ü—Ä–æ—Å—Ç–æ–µ —Ä–µ—Ö—ç—à–∏—Ä–æ–≤–∞–Ω–∏–µ
2 –†–µ—Ö—ç—à–∏—Ä–æ–≤–∞–Ω–∏–µ —Å –ø–æ–º–æ—â—å—é –ø—Å–µ–≤–¥–æ—Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª

```python
import time
import random

class HashTable:
    def __init__(self, size):
        self.size = size
        self.slots = [None] * self.size
        self.count = 0 # –∫–æ–ª–≤–æ

    def hash_function(self, key): #index
        return key % self.size

    def rehash(self, old_hash, step_size=1):
        return (old_hash + step_size) % self.size
    
    def is_full(self):
        return self.count == self.size
    
    def add(self, key, gen):
        if self.is_full():
            print("–•–µ—à-—Ç–∞–±–ª–∏—Ü–∞ –ø–µ—Ä–µ–ø–æ–ª–Ω–µ–Ω–∞")
            return
        
        hash_value = self.hash_function(key) #–≤—ã—á —Ö—ç—à

        if self.slots[hash_value] is None:
            self.slots[hash_value] = (key, gen)
            self.count += 1
        else:
            next_slot = self.rehash(hash_value)
            while self.slots[next_slot] is not None and self.slots[next_slot][0] != key:
                next_slot = self.rehash(next_slot)

            if self.slots[next_slot] is None:
                self.slots[next_slot] = (key, gen)
                self.count += 1
            else:
                self.slots[next_slot] = (key, gen)
                print(f"–ó–Ω–∞—á–µ–Ω–∏–µ –ø–æ –∫–ª—é—á—É {key} –æ–±–Ω–æ–≤–ª–µ–Ω–æ.")

    def get(self, key):
        start_slot = self.hash_function(key)

        x = '–ó–Ω–∞—á–µ–Ω–∏–µ –Ω–µ –Ω–∞–π–¥–µ–Ω–æ'
        stop = False
        found = False #–Ω–∞–π–¥–µ–Ω
        position = start_slot

        while self.slots[position] is not None and not found and not stop:
            if self.slots[position][0] == key:
                found = True
                x = self.slots[position][1]
            else:
                position = self.rehash(position)
                if position == start_slot:
                    stop = True

        return x

def remove(self, key):
        start_slot = self.hash_function(key)
        position = start_slot

        while self.slots[position] is not None:
            if self.slots[position][0] == key:
                self.slots[position] = None
                self.count -= 1
                print(f"–≠–ª–µ–º–µ–Ω—Ç —Å –∫–ª—é—á–æ–º {key} —É–¥–∞–ª—ë–Ω.")
                return
            position = self.rehash(position)
            if position == start_slot:
                break
        print("–≠–ª–µ–º–µ–Ω—Ç –Ω–µ –Ω–∞–π–¥–µ–Ω –¥–ª—è —É–¥–∞–ª–µ–Ω–∏—è.")

Size = int(input("–í–≤–µ–¥–∏—Ç–µ —Ä–∞–∑–º–µ—Ä —Ö—ç—à —Ç–∞–±–ª–∏—Ü—ã: "))
h_table = HashTable(Size) 
while True:
    command = input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–º–∞–Ω–¥—É (add/get/remove/show/exit): ")
    if command == 'add': 
        keyP = int(input("–í–≤–µ–¥–∏—Ç–µ –∫–ª—é—á: "))
        dataP = input("–í–≤–µ–¥–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ: ")
        h_table.add(keyP, dataP)
    elif command == 'get': 
        keyG = int(input("–í–≤–µ–¥–∏—Ç–µ –∫–ª—é—á –¥–ª—è –ø–æ–∏—Å–∫–∞: "))
        start_time = time.time()
        print(h_table.get(keyG))
        print("--- {0} ms ---".format(round((time.time() - start_time) * 1000)))
    elif command == 'remove':
        key = int(input("–í–≤–µ–¥–∏—Ç–µ –∫–ª—é—á –¥–ª—è —É–¥–∞–ª–µ–Ω–∏—è: "))
        h_table.remove(key)
    elif command == 'show': # –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ —Ö—ç—à —Ç–∞–±–ª–∏—Ü—ã
        print(h_table.slots)
    elif command == 'exit': # –ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ –ø—Ä–æ–≥—Ä–∞–º–º—ã
        break
    else:
        print("–ù–µ–≤–µ—Ä–Ω–∞—è –∫–æ–º–∞–Ω–¥–∞. –ü–æ–ø—Ä–æ–±—É–π—Ç–µ —Å–Ω–æ–≤–∞.")
```

```python
import time 
import random


class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for i in range(size)]

    def hash_function(self, key):
        return key % self.size ## index
    
    def add(self, key, value):
        hash_value = self.hash_function(key) #–≤—ã—á —Ö—ç—à
        self.table[hash_value].append((key, value)) 

    def remove(self, key):
        hash_value = self.hash_function(key)
        for i, (k, v) in enumerate(self.table[hash_value]):
            if k == key:
                del self.table[hash_value][i]  
                print(f"–≠–ª–µ–º–µ–Ω—Ç —Å –∫–ª—é—á–æ–º {key} —É–¥–∞–ª—ë–Ω.")
                return
        print(f"–≠–ª–µ–º–µ–Ω—Ç —Å –∫–ª—é—á–æ–º {key} –Ω–µ –Ω–∞–π–¥–µ–Ω –¥–ª—è —É–¥–∞–ª–µ–Ω–∏—è.")

    def get(self, key):
        start_slot = self.hash_function(key)

        for k1, k2 in self.table[start_slot]:
            if k1 == key:
                return k2
        return '–∑–Ω–∞—á–µ–Ω–∏–µ –Ω–µ –Ω–∞–π–¥–µ–Ω–æ'

Size = int(input("–í–≤–µ–¥–∏—Ç–µ —Ä–∞–∑–º–µ—Ä —Ö—ç—à —Ç–∞–±–ª–∏—Ü—ã: "))
h_table = HashTable(Size) 
while True:
    command = input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–º–∞–Ω–¥—É (add/get/remove/show/exit): ")
    if command == 'add': 
        keyP = int(input("–í–≤–µ–¥–∏—Ç–µ –∫–ª—é—á: "))
        dataP = input("–í–≤–µ–¥–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ: ")
        h_table.add(keyP, dataP)
    elif command == 'get': 
        keyG = int(input("–í–≤–µ–¥–∏—Ç–µ –∫–ª—é—á –¥–ª—è –ø–æ–∏—Å–∫–∞: "))
        start_time = time.time()
        print(h_table.get(keyG))
        print("--- {0} ms ---".format(round((time.time() - start_time) * 1000)))
    elif command == 'remove':
        keyR = int(input("–í–≤–µ–¥–∏—Ç–µ –∫–ª—é—á –¥–ª—è —É–¥–∞–ª–µ–Ω–∏—è: "))
        h_table.remove(keyR)
    elif command == 'show': # –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ —Ö—ç—à —Ç–∞–±–ª–∏—Ü—ã
        print(h_table.table)
    elif command == 'exit': # –ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ –ø—Ä–æ–≥—Ä–∞–º–º—ã
        break
    else:
        print("–ù–µ–≤–µ—Ä–Ω–∞—è –∫–æ–º–∞–Ω–¥–∞. –ü–æ–ø—Ä–æ–±—É–π—Ç–µ —Å–Ω–æ–≤–∞.")
```

–ù–∞–ø–∏—Å–∞—Ç—å –ø—Ä–æ–≥—Ä–∞–º–º—É,  –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç —Ö–æ—Ç—è –±—ã –æ–¥–∏–Ω —Å–ø–æ—Å–æ–± —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á.

```python
def is_valid(row, col, board): 
    # –ø—Ä–æ–≤–µ—Ä–∫–∞, –±—å—é—Ç—Å—è –ª–∏ —Ñ–µ—Ä–∑–∏ –≤ —Ç–æ–π –∂–µ —Å—Ç—Ä–æ–∫–µ –∏–ª–∏ –¥–∏–∞–≥–æ–Ω–∞–ª—è—Ö 
    for i in range(row): 
        if board[i] == col or board[i] - i == col - row or board[i] + i == col + row: 
            return False 
    return True 
def solve(size, row, board, results): 
    # –µ—Å–ª–∏ –≤—Å–µ —Å—Ç–æ–ª–±—Ü—ã –∑–∞–ø–æ–ª–Ω–µ–Ω—ã, –¥–æ–±–∞–≤–ª—è–µ–º —Ä–∞—Å—Å—Ç–∞–Ω–æ–≤–∫—É —Ñ–µ—Ä–∑–µ–π –≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã 
    if row == size: 
        results.append(board.copy()) 
        return 
    # –ø—ã—Ç–∞–µ–º—Å—è –ø–æ—Å—Ç–∞–≤–∏—Ç—å —Ñ–µ—Ä–∑—è –≤ –∫–∞–∂–¥—É—é —Å—Ç—Ä–æ–∫—É —Ç–µ–∫—É—â–µ–≥–æ —Å—Ç–æ–ª–±—Ü–∞ 
    for col in range(size): 
        if is_valid(row, col, board): 
            board[row] = col 
            solve(size, row+1, board, results) 
            board[row] = -1 
def find_all_solutions(size): 
    results = [] 
    board = [-1] * size #–¥–æ—Å–∫–∞
    solve(size, 0, board, results) 
    return results 

def print_solution(solution):
    size = len(solution)
    for row in range(size):
        line = ""
        for col in range(size):
            if solution[row] == col:
                line += "Q "
            else:
                line += ". "
        print(line)
    print()

all_solutions = find_all_solutions(8) 
print(f"–ù–∞–π–¥–µ–Ω–æ {len(all_solutions)} —Ä–µ—à–µ–Ω–∏–π –¥–ª—è –¥–æ—Å–∫–∏ 8x8")
for i, solution in enumerate(all_solutions[:2], 1):
    print(f"–†–µ—à–µ–Ω–∏–µ #{i}:")
    print_solution(solution)
```
